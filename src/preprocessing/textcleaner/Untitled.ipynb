```python
def get_readability_mark(score):
    score = score * 100
    if score in range(0, 20):
        return 'E'
    elif score in range(21, 40):
        return 'D'
    elif score in range(41, 60):
        return 'C'
    elif score in range(61, 80):
        return 'B'
    else:
        return 'A'
```

```python
re_nonalpha = re.compile(r'[^a-zA-Z\sàâäéèêëïîôöùûüÿç]')
re_elision = re.compile(r"(c|l|d|m|n|s|t|qu)(’|')")
re_initiale = re.compile(r"[A-Z]\. ")
re_digit = re.compile(r'\d')
re_acronyms = re.compile(r' ([A-Z]\.)+( |$)')

def clean_text(text):
    for rule in [re_acronyms, re_initiale, re_elision, re_digit, re_nonalpha]:
        text = re.sub(rule, '', text)
    text = text.lower()
    text = text.replace(' ', '') 
    return text
```

```python
icdar_text = clean_text(icdar_text)
icdar_text[:100]
```

```python
def to_ngram_set(text, n=3):
    grams = ngrams(text, n)
    list_grams = [''.join(g) for g in grams]
    return list_grams
```

```python
n = 3
list_grams = to_ngram_set(icdar_text, n=n)
```

```python
dic_gram = dict(Counter(list_grams))
filter_dic_gram = {x:y for x, y in dic_gram.items() if y > 100}
```

```python
with open(f'data/fr/{n}_gram_dic.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(filter_dic_gram))
```
