## GS dictionnary generation

```python
icdarpath = 'ICDAR2017_datasetPostOCR_v1.2/ICDAR2017_datasetPostOCR_Full_12M_v1.2/'
```

```python
text_l = []
for doc in glob(f'{icdarpath}/fr_*/**/*.txt', recursive=True):
    with open(doc, encoding='utf-8') as f:
        text = f.read()
    textlines = text.split('\n')
    gstext = textlines[-1]
    gstext = gstext.replace('[ GS_aligned]', '')
    gstext = gstext.replace('#', '')
    gstext = gstext.replace('@', '')
    gstext = gstext.strip()
    text_l.append(gstext)
    # print(gstext)
    # break
```

```python
with open('data/gs_icdar.txt', 'w', encoding='utf-8') as f:
    gs = ' '.join(text_l)
    f.write(gs)
```

```python
with open('data/gs_icdar.txt', encoding='utf-8') as f:
    gs = f.read()
```

```python
from itertools import tee, islice
from collections import Counter
def ngrams(lst, n):
    tlst = lst
    while True:
        a, b = tee(tlst)
        l = tuple(islice(a, n))
        if len(l) == n:
          yield l
          next(b)
          tlst = b
        else:
          break
```

```python
# unigrams
gs_words = re.findall(r'\w+', gs)
unigrams_counts = dict(Counter(ngrams(gs_words, 1)))
dic = {}
for k, v in unigrams_counts.items():
    
    if len(k) > 1:
        index = ' '.join(k)
    else:
        index = k[0]
    dic[index] = v
    # dic = dic + f'{index} {v}\n'
```

```python
# deletes 1 letter entries, except few exceptions
l_del = []
for k, v in dic.items():
    if len(k) == 1:
        if k not in ('a', 'à', "y", 'ô', 'M'):
            l_del.append(k)
for d in l_del:
    del dic[d]
```

```python
txt_dic = ''
for k, v in dic.items():
    txt_dic = txt_dic + f'{k} {v}\n'
txt_dic
```

```python
with open('data/fr_dic.txt', 'w', encoding='utf-8') as f:
    f.write(txt_dic)
```
